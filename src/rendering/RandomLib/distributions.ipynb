```python
import numpy as np
def sample_trunc_norm(mu, sigma, a, b):
    x = None
    while((x is None) or (x < a) or (x > b)):
        x = np.random.normal(mu, sigma)
    return x
```

```python
n = 10000
mu = 0.0
sigma = 50.0
a = -180.0
b = 180.0
y = np.zeros((n,))
for i in range(n):
    y[i] = sample_trunc_norm(mu, sigma, a, b)
```

```python
import matplotlib.pyplot as plt
plt.hist(y, bins=50)
plt.xlim([-180,180])
plt.show()
```

```python
def random_shell_coord(radius, phi_sigma):
    theta = np.deg2rad(np.random.uniform(0.0, 360.0))
    phi = np.deg2rad(sample_trunc_norm(90.0, phi_sigma, 0.0, 180.0))
    x = radius * np.cos(theta) * np.sin(phi)
    y = radius * np.sin(theta) * np.sin(phi)
    z = radius * np.cos(phi)
    return x, y, z
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
n = 1000
X = np.zeros((n,))
Y = np.zeros((n,))
Z = np.zeros((n,))
for i in range(n):
    X[i],Y[i],Z[i] = random_shell_coord(1.0, 20.0)
```

```python
ax.scatter(X,Y,zs=Z)
ax.set_zlim([-1,1])
ax.set_ylim([-1,1])
ax.set_xlim([-1,1])
plt.show()
```
